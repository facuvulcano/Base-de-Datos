- SGBD como alternativa a los archivos individuales.
- Los SGBD es un conjunto de programas que gestiona y controla la creacion, manipulacion y acceso a la base de datos.
- Brinda un nivel de abstraccion al desarrollador, al desacoplar al software de aplicacion del problema de almacenamiento y acceso a los datos. Esto resuelve el problema conocido como dependencia de datos.
- Independencia de datos: Es la propiedad del SGBD consistente en que cambios en la estructura de la base de datos no repercutan en los programas o sistemas de informacon que la utilizan.

Su funcion principal es servir de interfaz entre los usuarios y las bases de datos, facilitando las siguientes tareas:

    1- Almacenamiento y Consulta: Ofrecer estructuras eficientes. Ofrecer un lenguaje de consulta (aumenta la productividad)

    2- Definicion de Datos: Permite definir la estructura de la base de datos, incluyendo las tablas, los campos y las relaciones entre ellos.

    3- Manipulacon de datos: Permite realizar operaciones como insertar, actualizar, elominar y consultar datos dentro de la base de datos mediante lenguajes de manipluacion de datos como SQL.

    4- Control de Acceso: Gestiona la seguridad y el acceso a los datos, asegurando que solo los usuarios autorizaods pueden acceder o modficiar la informaacion.

    5- Integridad de Datos: Mantiene la integridad de los datos asegurando que se cumplan ciertas reglas y restricciones predefinidas.

    6- Administracion de Transacciones: Maneja las transacciones de la base de datos, asegurando que las operaciones se realicen de manera completa y precisa, y manteniendo la coherencia de los datos en casos de fallos. Implicia la posibilidad de volver atras (rollback) los cambios

    7- Respaldo y Recuperacion: Permite realizar copias de seguridad y restaurar datos en caso de perdida o corrupucion.

    8- Desempenio: Optimiza el rendimiento de las consultas y operaciones en la base de datos para asegurar un acceso rapido y eficiente a los datos.

    9- Concurrencia: Permitir el acceso en simultaneo de muchos usuarios

Algunos ejemplos de SGBD son MySQL, PostgreSQL, Oracle, Microsoft SQL Server y SQLite. Estos sistemas se utilizan en una variedad de aplicaciones, desde pequenias aplicacipnes de escritorio hasta grandes sistemas empresariales y aplicaciones web.

Arquitectura de 3 capas ANSI/SPARC

- El ANSI-SPARC propuso en 1975 una arquitectura en 3 niveles de abstraccion para la descripcion/representacion de los datos de una base de datos.


- Modelo interno: Representa la forma en que los datos se almacenan utilizando estructuras de datos y organizaciones de archivos. Representa como perciben los datos el sistema operativo y el SGBD.

- Modelo conceptual: Describe la semantica de los datos, abstrayendose de su implementacion fisica. Describe entidades, tipos de datos, operaciones y restricciones de seguridad y de integridad.

- Modelo externo: Representa la forma en que los usuarios perciben los datos.

Esta arquitectura asegura la independencia de datos, tanto fisica como logica.

Las bases de datos tradicionales almacenan datos que pueden enunciarse a traves de proposiciones. Un modelo de datos debe incluir los siguientes elementos:
    - Un conjunto de objetos, sus propiedades e interrelaciones entre ellos, que representa la estructura.
    - Un conjunto de operaciones, o lenguaje, que permite manipular los datos.
    - Restricciones sobre los objetos, las interrelaciones y las operaciones.

Modelo Entidad-Interrelacion

En 1976 P.Chen propuso un modelo esquematico basado en entidades, interrelaciones y atributos para el modelado semantico de datos. Un modelo de datos entidad-interrelacion es una tecnica de modelado utilizada en ingeneiria de software y bases de datos para representar y estructurar los datos de un sistema de informacion. Este modelo se centra en identificar las entidades relevantes, las relaciones entre ellas y los atributos que describen estas entidades y relaciones.

1- Tipo de Entidad: tipo de objeto en particular que queremos almacenar en nuestra base. 
- En el diagrama E-R, una enitdad se representa como un rectangulo con el nombre de la entidad dentro.
- Ejemplos: Futbolista, pais, cancion. Se los recuadra a estos.
- Una entidad es una instancia de un tipo de entidad.
    ej: Juan Fernando Quintero (Futbolista), Colombia (Pais).

2- Atributo: Es una propiedad que describe a la entidad:
- Representacion: En el diagrama E-R, los atributos se representan como ovalos conectados a sus respectivas entidades mediante una linea.
- Ejemplos: Futbolista --> Cotizacion, Pais --> Poblacion, Cancion --> Duracion. Estas propiedades se marcan con un circulo debajo del tipo de entidad.

3- Tipo de interrelacion: Es la definicion de un conjunto de relaciones o asociaciones similares entre dos o mas tipos de entidades.
- Representacion: En el diagrama E-R, una relacion se representa como un rombo con el nombre de la relacion dentro, y se conecta a las entidades participantes mediante lineas.
- Ejemplo: Futbolista nacio en Pais
- Una interrelacion en concreto es: Didier Drogba nacio en Costa de Marfil.

Atributos:

Cada entidad tendra valores particulares para cada uno de los atributos del tipo de entidad al que corresponde
- Los representaremos de la siguiente forma: Ejemplo para el tipo de entidad Pais (nombre, poblacion, superficie):
    - (Argentina, 46.044.703, 2.780.400 km2)

El dominio de un atributo es el conjunto de valores que el mismo puede tomar
- Para la entidad Pais:
    - El atributo nombre es una cadena de caracteres (String)
    - El atributo poblacion es un numero entero positivo
    - El atributo superficie es un numero real positivo

En ciertos casos puede permitirse que un atributo de un tipo de entidad no tome ningun valor concreto en ciertas instancias. En estos casos diremos que el atributo toma un vaor nulo o NULL.

- Esta situacion se da en general por alguno de los siguientes motivos:
    - La entidad particular no posee aun ese atributo (Ejemplo: un futbolista que nunca fue cotizado)
    - El valor de ese atributo para esa entidad se desconoce, aunque existe. (Ejemplo: en un pais en que no hay censos de poblacion).

Atributos compuestos vs simples

- Atributo simple:
    - Definicion: Un atributo simple es un atributo que no se puede descomponer en subcomponentes mas pequenios. Es indivisible en terminos de la informacion que representa.
    - Ejemplo:
        - "Nombre" puede ser considerado un atributo simple si se almacena como un solo campo son descomponerse en "Nombre de pila" y "Apellido"
        - "Numero de telefono" si se guarda en un solo campo y no se separa en codigo de pais, codigo de area y numero de suscriptor
    - Ventajas:
        - Simplicidad: Facilita el almacenamiento y la recuperacion de informacion
        - Facilidad de uso: Mas facil de manejar en temrinos de consiltas y actualizaciones
    
- Atributos Compuestos:
    - Definicion: Un atributo compuesto es un atributo que se puede descomponer en subcomponentes mas pequenios que tienen significados independientes.
    - Ejemplo:
        - "Nombre completo" puede descomponerse en "Nombre de pila" y "Apellido"
        - "Direccion" puede ser un atributo compuesto que se divide en "Calle", "Numero", "Ciudad", "Codigo Postal", y "Pais".
    - Ventajas:
        - Flexbilidad: Permite un analisis mas detallado y especifigo de la informacion
        - Normalizacion: Facilita la normalizacion de la base de datos, lo que puede reducir redundancias y mejorar la integridad de los datos.

Atributos multivaluados vs monovaluados

- Atributos Monovaluados:
    - Definicion: Un atributo monovaluado es un atributo que solo puede tener un unico valor para una entidad en un momento dado.
    - Ejemplo:
        - "Numero de seguro social" de una persona
        - "Fecha de nacimiento" de una persona
        - "ID de producto"

- Atributos Multivaluados:
    - Definicion: Un atributo multivaluado es un atributo que puede tener mas de un valor para una entidad en un momento dado.
    - Ejemplo:
        - "Numero de telefono" de una persona (una persona puede tener varios numeros de telefono)
        - "Habilidades" de un empelado
        - "Correos electronicos"

Atributos almacenados vs derivados

- Atributos Almacenados:
    - Definicion: Un atributo almacenado es un atributo cuyo valor se guarda explicitamente en la base de datos.
    - Ejemplo:
        - "Nombre" de una persona
        - "Fecha de nacimiento"
        - "precio" de producto
    
- Atributos derivados:
    - Definicion: Un atributo derivado es un atributo cuyo valor no se guarda directamente en la base de datos, sino que se calcula a partir de otros atributos almacenados cuando se necesita.
    - Ejemplo:
        - "Edad" puede derivarse a partir de la fecha de nacimiento y la fecha actual
        - "Monto total" de un pedido, puede calcularse sumando los precios de los productos incluidos en el pedido.


Entidades

- Conjunto de entidades: Al conjunto de ocurrencias o instancias de un determinado tipo de entidad en un estado determinado de la base de datos se lo denomina conjunto de entidades de ese tipo de entidad. 
    - Por ejemplo, para el tipo de entidad Pais, nuestra base de datos podria tener cargdas en un momento dado las siguientes instancias:
        - (Argentina, 46.000.000, 2.700.000km2)
        - (Costa de marfil, ....., .....)
        - (Serbia y Montenegro, ...., ....)
    
- Restriccion de unicidad: 
    - Todo tipo de entidad debe tener un subconjunto de atributos cuyos valores sean necesariamente distintos para cada instancia de entidad que puede existir.
    - Dichos atributos se llaman atributos clave o identificadores unicos.
    - Si no los encontramos, debemos crear uno (id).
    - Al ser distintos para cada entidad, permite identificarlas.
    - En el diagrama Entidad-Interrelacion los representamos subrayados.
    - El conjunto de atributos clave debe ser minimal, es decir, ningun subconjunto del mismo debe ser capaz de identificar univocamente a las entidades.
    - Aun asi, es posible que exista mas de un conjunto de atrivutos clave para un tipo de entidad.
    - La restriccion de unicidad para los valores de los atributos clave no es una propeidad del conjunto de entidades en el estado actual de la base de datos, sino que es una restriccion respecto a todos los estados posibles de la misma.

Interrelacioens

- Aridad: La aridad o grado de un tipo de interrelacion es la cantidad de tipos de entidad que cparticion del mismo

- Cardinalidad: Es la maxima cantidad de instancias de cada tipo de entidad que pueden relacionarse con una instancia concreta de los tipos de entidades restantes. En interrelaciones binarias las cardinalidades posibles son: 1:1, 1:N, N:1 y M:N.

Cardinalidades:

1- Relacion uno a uno (1:1)

- Ejemplo: Una persona tiene un numero de seguro social

- Lectura: Cada cantidad del conjunto A se relaciona con una y sola una entidad del conjunto B y viceversa.

- Representacion: Se coloca un 1 cerca de ambos lados de la relacion entre las entidades A y B.

2- Relacion uno a muchos (1:N):

- Ejemplo: Un profesor imparte varias clases, pero cada clase es impartida por un solo profesor.

- Lectura: Cada entidad del conjunto A puede relacionarse con una o mas entidades del conjunto B, pero cada entidad de conjunto B se relaciona con una y solo una entidad del conjunto A.

- Representacion: Se colocia un 1 en el lado de la entidad A y un N en el lado de la entidad B.

3-  Relación Muchos a Uno (N:1)

- Ejemplo: Muchos estudiantes se inscriben en un solo curso, pero cada curso puede tener muchos estudiantes.

- Lectura: Es la inversa de la relación uno a muchos: "Cada entidad del conjunto A puede relacionarse con una y solo una entidad del conjunto B, pero cada entidad del conjunto B puede relacionarse con una o más entidades del conjunto A."

- Representación: Se coloca un "N" en el lado de la entidad A y un "1" en el lado de la entidad B.

4- Relación Muchos a Muchos (M:N)

- Ejemplo: Los estudiantes pueden inscribirse en múltiples cursos, y cada curso puede tener múltiples estudiantes.

- Lectura: "Cada entidad del conjunto A puede relacionarse con una o más entidades del conjunto B, y cada entidad del conjunto B puede relacionarse con una o más entidades del conjunto A."
- Representación: Se coloca un "M" y un "N" en los lados correspondientes de la relación entre A y B.

5- Cardinalidades Mínimas y Máximas

- Ejemplo: Si se indica que un profesor imparte mínimo 1 y máximo 4 clases, se especifica la cardinalidad en ambos lados de la relación.

- Lectura: "Cada entidad del conjunto A está obligada a relacionarse con al menos una entidad del conjunto B, pero no más de cuatro entidades."

- Representación: Generalmente se usan notaciones como (1,4) o (0,1) para denotar las cardinalidades mínimas y máximas.

6- Notaciones de Cardinalidad

- Notación de Chen: Es la clásica representación gráfica con un rombo que conecta dos entidades, donde se especifican los números correspondientes a las cardinalidades.

- Notación de Crowsfoot: Usa símbolos como patas de gallo para indicar "muchos" y una línea para indicar "uno".

1- Relación:

- En una base de datos relacional, una "relación" se refiere a una tabla. Es una estructura que organiza datos en filas y columnas. Cada fila (o "tupla") representa un registro individual, y cada columna (o "atributo") representa un campo de ese registro.

- Por ejemplo, si tienes una tabla llamada Persona, cada fila de esa tabla representaría a una persona con atributos como nombre, DNI, género, etc.

2- PK (Clave Primaria o Primary Key)

- La Clave Primaria es un campo o conjunto de campos que identifican de manera única cada registro en una tabla. Ningún valor en la clave primaria puede ser NULL, y no puede haber dos registros con el mismo valor en la clave primaria.

- Ejemplo: En una tabla Persona, el DNI podría ser la clave primaria porque cada persona tiene un DNI único.

3- CK (Clave Candidata o Candidate Key)

- Una Clave Candidata es cualquier conjunto de columnas que puede actuar como una clave primaria porque identifica de manera única a cada fila en una tabla. En una tabla puede haber más de una clave candidata, pero solo una de ellas se selecciona como clave primaria.

- Ejemplo: Si en una tabla Persona además del DNI, cada persona tuviera un número de pasaporte único, tanto el DNI como el número de pasaporte serían claves candidatas. Sin embargo, solo una (digamos, DNI) se elegirá como la clave primaria.

4- FK (Clave Foránea o Foreign Key)

- La Clave Foránea es un campo o conjunto de campos en una tabla que se utiliza para establecer una relación con otra tabla. Este campo hace referencia a la clave primaria de otra tabla, creando una relación entre las dos tablas. La clave foránea asegura que el valor en ese campo coincida con un valor existente en la clave primaria de la tabla referenciada, o que sea NULL si se permite.

- Ejemplo: En una tabla Matrimonio, el DNI_Esposo1 y DNI_Esposo2 podrían ser claves foráneas que hacen referencia al DNI en la tabla Persona. Esto asegura que ambos cónyuges registrados en el matrimonio existan en la tabla Persona.

Resumen de cómo se conectan:

- Una Relación es básicamente una tabla en una base de datos.

- La PK es el identificador único de una tabla, que distingue cada registro.

- Las CK son todos los posibles identificadores únicos (columnas o combinaciones de columnas) de una tabla.

- La FK establece una conexión entre dos tablas, asegurando que los datos en una tabla correspondan a los datos en otra.
